## 微积分总览
### 微积分是什么
微积分是函数一和函数二之间的桥梁

```python
import numpy as np
import matplotlib.pyplot as plt

#x = [0.1*i for i in range(0,51)]
#y = [0.01*i*i for i in range(0,51)]

x_1 = [0.1*i for i in range(0,51)]
y_1 = [0.2*i for i in range(0,51)]

x_2 = [0.1*i for i in range(0,51)]
y_2 = [2 for i in range(0,51)]

plt.figure(figsize = (10,6))

ax = plt.subplot(1,2,1)
ax.plot(x_1,y_1)
ax.set_title("distance vs time")
ax.set_xlabel("time")
ax.set_ylabel("distance")

ax = plt.subplot(1,2,2)
ax.plot(x_2,y_2)
ax.set_title("speed vs time")
ax.set_xlabel("time")
ax.set_ylabel("speed")
```

在上面的图中，左边是函数一，右边是函数二。函数一代表这距离和时间的关系，函数二代表这速度和时间的关系。  
从函数一到函数二的过程叫做微分。    
从函数二到函数一的过程叫做积分。    
假如我们设置函数一为$f(t)$ ,t是时间变化量也是自变量，f(t)是随着时间变化距离的变换量也是因变量   
那么函数二就变成了$\frac{df}{dt}$
通过上面的函数可以很直接的看到，我们需要使用$s=vt或者v=\frac{s}{t}$这两个公式就可以得到这两个函数的变换 
当然，代数当然可以在两个函数之间进行转换    
现在稍微增加一点难度计算2-3s是的速度，这也很好算,只需要计算出距离差和时间差就可以，然后使用上面的公式进行带入就可以  

```python
x = [0.1*i for i in range(0,51)]
y = [0.01*i*i for i in range(0,51)]

x_1 = [0.1*i for i in range(0,51)]
y_1 = [0.2*i for i in range(0,51)]


plt.figure(figsize = (10,6))

ax = plt.subplot(1,2,1)
ax.plot(x,y)
ax.set_title("distance vs time")
ax.set_xlabel("time")
ax.set_ylabel("distance")

ax = plt.subplot(1,2,2)
ax.plot(x_1,y_1)
ax.set_title("speed vs time")
ax.set_xlabel("time")
ax.set_ylabel("speed")
```

这个函数一和上述函数一很像只不过直观应该是一个加速的过程，因为速度在不断的变大，通过这两个例子可以看出函数二的要点就是它代表这函数一的变化率    
就像假如我们知道函数二$s=at$,我们能不能求出函数一，这里我们从速度回到距离，其实就是积分，我们通过计算函数二的面积皆可以得到对应的函数一的距离。 
三角形的面积公式为二分之一的底乘高，所以我们可以得到面积的表达式$\frac{1}{2}at^2$。微积分告诉我们这就是函数一
