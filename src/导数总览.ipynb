## 导数总览
导数的介绍和计算    
核心思想：函数都是成对的，比如距离和速度，微积分就是讨论两种函数之间关系的学科      
### 三个重要的函数
$y=x^n,\frac{dy}{dx}=nx^{n-1}$     
$y=sinx,\frac{dy}{dx}=cosx$    
$y=e^x,\frac{dy}{dx}=e^x$  

例子，现在飞机有两块记录表，一块用来计算高度，一块用来计算速度，很不幸，计算速度的那块记录表丢失了，我们应该怎么通过第一块记录表来推到出第二块记录表呢  
如果第一块里程表仅仅有一小时、两小时或者三小时之后的读数，这样微积分可就帮不上什么忙了，因为只知道一个小时之后的距离，那样仅仅能算出一个小时之内平均速度，不能算出来什么时候减速什么时候加速，平均很简单、平均不需要微积分。微积分关心的是瞬时、即任意时刻的情况

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = [0.1*i for i in range(0,51)]
y = [0.01*i*i for i in range(0,51)]



plt.figure(figsize = (10,6))

ax = plt.subplot(1,2,1)
ax.plot(x,y)
ax.set_title("distance vs time")
ax.set_xlabel("time")
ax.set_ylabel("distance")
```

ok，目前为止，还只是代数运算    
到目前为止，仍然是小范围的平均值，还是不瞬时的情形

$\frac{dy}{dx} = lim \frac{\Delta y}{\Delta x}$
